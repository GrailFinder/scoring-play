```python
%matplotlib inline

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df_train = pd.DataFrame.from_csv('train.csv')
df_test = pd.DataFrame.from_csv('test.csv')

df_train.head()
```

```python
def group_by_feature(df, feat):
    columns = df.columns.values
    for col in columns:
        if col != feat:
            #print(df[[col, feat]].groupby([col]).mean())
            try:
                df[[col, feat]].groupby([col]).mean().plot()
            except Exception as e:
                print(e)
            
group_by_feature(df_train, 'SalePrice')
```
